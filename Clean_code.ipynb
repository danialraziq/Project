```python

from google.colab import files
import re
```

```python
fil = files.upload()
```

```python
peranggan ="Hi, saya danial, dani suka buah"
nama_ulang = re.compile(r"D\anial")
nama_ulang.findall(peranggan)
nama_ulang.sub("",peranggan)
```

```python
kunci = list(fil.keys())
text = str(fil[kunci[0]])

print(type(text))
text
```

```python
#replace patern
#\\n\\n5\\n00:19.270 --> 00:20.530\\n

gantiAyat = r"\\n\\n\d{1,3}\\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\\n"

gantu_masa = re.compile(gantiAyat)
gantu_masa.findall(text)

ayatBaharu = gantu_masa.sub(" ",text)
ayatBaharu[9:-1]
```

```python
len(ayatBaharu)
word = ayatBaharu.split()

print(type(word))
print(len(word))

for wordi in range(len(word)):
  if len(word[wordi]) == 4 and not word[wordi][-1]==".":
    word[wordi] = word[wordi][0] + "***"
word
```

```python
clean_text = " ".join(word)
```

```python
with open("clean_code.txt", "w") as f:
  f.write(clean_text)

files.download("clean_code.txt")
```
